```python
def plot_3d_func(func, min, max, n=100):
    xgrid = np.linspace(min, max, n)
    ygrid = xgrid

    xmat, ymat = np.meshgrid(xgrid, ygrid, indexing="ij")

    zmat = func(xmat, ymat)

    ax = plt.axes(projection="3d")
    ax.plot_surface(xmat, ymat, zmat, cmap="viridis")
    ax.set_title("surface")
    ax.set_xlabel("m")
    ax.set_ylabel("n")
    ax.set_zlabel("f")
    plt.show()
```

```python
plot_3d_func(agent.solution[0].deposit_stage.dFunc, 0, 10)
```

```python
plot_3d_func(agent.solution[0].deposit_stage.cFunc, 0, 10)
```

```python
from scipy.interpolate import LinearNDInterpolator
```

```python
d = agent.solution[0].deposit_stage.dFunc.f_values
m = agent.solution[0].deposit_stage.dFunc.x_values
n = agent.solution[0].deposit_stage.dFunc.y_values
```

```python
idx = np.isnan(n)
n = n[~idx]
d = d[~idx]
m = m[~idx]
```

```python
X = np.linspace(0, 10, 1000)
X, Y = np.meshgrid(X, X, indexing="ij")
```

```python
interp = LinearNDInterpolator(list(zip(m, n)), d, rescale=True)
Z = np.maximum(0.0, interp(X, Y))
Z = np.nan_to_num(Z)
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis")
ax.set_title("surface")
ax.set_xlabel("m")
ax.set_ylabel("n")
ax.set_zlabel("f")
plt.show()
```
